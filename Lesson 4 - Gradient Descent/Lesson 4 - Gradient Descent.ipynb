```python
import numpy as np

def gradient_descent(x, y):
#     print(f"x {y} y {x}")
    m_current = b_current = 0
    iterations = 1000
    xlength = len(x)
    learning_rate = 0.083

#     print('m_current %2d b_current %2d' % (m_current, b_current))
    for i in range(iterations):
        y_predicted = m_current * x + b_current
        mse = (1/xlength) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/xlength) * sum(x * (y-y_predicted))
        bd = -(2/xlength) * sum(y - y_predicted)
        m_current = m_current - learning_rate * md
        b_current = b_current - learning_rate * bd
        print ("iteration {}, m {}, b {}, mse {}".format(i, m_current, b_current, mse))

x = np.array([1, 2, 3 , 4, 5])
y = np.array([5, 7, 9, 11, 13])

gradient_descent(x, y)
```
