```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plts
from sklearn.datasets import load_iris
from sklearn.svm import SVC

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df


```

```python
df['target'] = iris.target
df['flower_names'] = df.target.apply(lambda x: iris.target_names[x])
df
```

```python
setosaDf = df[df.target == 0]
versicolorDf = df[df.target == 1]
virginicaDf = df[df.target == 2]

plts.ylabel('sepal length (cm)')
plts.ylabel('sepal width (cm)')
plts.scatter(setosaDf['sepal length (cm)'], setosaDf['sepal width (cm)'], color='green', marker='+')
plts.scatter(versicolorDf['sepal length (cm)'], versicolorDf['sepal width (cm)'], color='blue', marker='.')
plts.scatter(virginicaDf['sepal length (cm)'], virginicaDf['sepal width (cm)'], color='orange', marker='x')
```

```python
X = df.drop(['target', 'flower_names'], axis=1)
y = df.target

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

linear_kernel_model = SVC(kernel='linear')
linear_kernel_model.fit(X_train, y_train)

regularization10_kernel_model = SVC(C=10)
regularization10_kernel_model.fit(X_train, y_train)

gamma10_kernel_model = SVC(gamma='auto')
gamma10_kernel_model.fit(X_train, y_train)


print(linear_kernel_model.score(X_test, y_test))
print(regularization10_kernel_model.score(X_test, y_test))
print(gamma10_kernel_model.score(X_test, y_test))

```
